**Question 4: What are some ways to prevent merge conflicts?**


Work in separate branches instead of directly on main.
Communicate with your teammates about when you plan to merge.
Pull frequently from the main branch to stay up to date.
For Jupyter Notebooks: try to synchronize runs and clean outputs before merging.
